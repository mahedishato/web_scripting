```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from bs4 import BeautifulSoup
import requests
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) Applewebkit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

webpage= requests.get('https://en.wikipedia.org/wiki/List_of_hospitals_in_Bangladesh', headers=header).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('li')[0].text
```

```python
hospital = []
for i in soup.find_all('li'):
    hospital.append(i.text.strip())
```

```python
len(hospital)
```

```python
df = pd.DataFrame(hospital)
```

```python
df.to_csv('hospital.csv')
```
